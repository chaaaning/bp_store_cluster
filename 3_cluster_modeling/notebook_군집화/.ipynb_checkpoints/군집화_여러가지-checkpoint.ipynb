```python
import pandas as pd
import numpy as np
import os
import warnings

def row_num(x): return pd.set_option('display.max_rows',x)
def col_num(x): return pd.set_option('display.max_columns',x)

row_num(10)
col_num(10)

warnings.simplefilter('ignore')

os.chdir(r'C:\\Users\\user\\bupyeonggu\\store_cluster\\data_and_notebook\\정제데이터')
df = pd.read_csv('GWR_회귀계수_영업.csv')
df_p = pd.read_csv('GWR_회귀계수_폐업.csv')
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering as agcluster
from sklearn.cluster import KMeans

%matplotlib inline
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(15,10)
plt.rcParams['figure.facecolor']='grey'
plt.rcParams['axes.grid']=True
```

# 1. 스케일링 데이터 셋 생성

```python
from sklearn.preprocessing import MinMaxScaler as MMS

tmp = pd.DataFrame(MMS().fit_transform(df.drop('BAS_ID',axis=1)), columns=df.drop('BAS_ID',axis=1).columns)
p_tmp = pd.DataFrame(MMS().fit_transform(df_p.drop('BAS_ID',axis=1)), columns=df_p.drop('BAS_ID',axis=1).columns)

tmp2 = pd.concat([df['BAS_ID'],tmp],axis=1)
p_tmp2 = pd.concat([df_p['BAS_ID'],p_tmp],axis=1)

df_scale = tmp2.copy()
df_pscale = p_tmp2.copy()

df_scale.drop(['x','y'],axis=1,inplace=True)
df_pscale.drop(['x','y'],axis=1,inplace=True)
```

# 2. 계층적 군집화 실행 (군집 5개 설정)

- Metric : Cosine, Euclidean
- Affinity : Ward, Complete, Single, Average

## 2.1. 스케일링하지 않은 데이터의 계층 군집화 (cluster=5)

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_c5 = df[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=5, affinity=aff, linkage=lng).fit(df.drop('BAS_ID',axis=1))
            col_name = 'O5_'+aff+'_'+lng
            df_c5[col_name]=cluster.labels_
        except:
            continue
```

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_p5 = df_p[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=5, affinity=aff, linkage=lng).fit(df_p.drop('BAS_ID',axis=1))
            col_name = 'C5_'+aff+'_'+lng
            df_p5[col_name]=cluster.labels_
        except:
            continue
```

## 2.2. 스케일링 데이터의 계층군집화 (cluster=5)

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_sc5 = df_scale[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=5, affinity=aff, linkage=lng).fit(df_scale.drop('BAS_ID',axis=1))
            col_name = 'O5s_'+aff+'_'+lng
            df_sc5[col_name]=cluster.labels_
        except:
            continue
```

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_sp5 = df_pscale[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=5, affinity=aff, linkage=lng).fit(df_pscale.drop('BAS_ID',axis=1))
            col_name = 'C5s_'+aff+'_'+lng
            df_sp5[col_name]=cluster.labels_
        except:
            continue
```

## 2.3. 스케일링 데이터 계층 군집화 (cluster=9)

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_sc9 = df_scale[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=9, affinity=aff, linkage=lng).fit(df_scale.drop('BAS_ID',axis=1))
            col_name = 'O9s_'+aff+'_'+lng
            df_sc9[col_name]=cluster.labels_
        except:
            continue
```

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_sp9 = df_pscale[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=9, affinity=aff, linkage=lng).fit(df_pscale.drop('BAS_ID',axis=1))
            col_name = 'C9s_'+aff+'_'+lng
            df_sp9[col_name]=cluster.labels_
        except:
            continue
```

# 3. 비지도 학습 군집화 실행 (K-means Clustering/군집 5개 설정)

```python
km_cluster= KMeans(n_clusters=5).fit(df_pscale.drop('BAS_ID',axis=1))
df_sp5['k_means_C5']=km_cluster.labels_

df_sp5.drop('BAS_AR', axis=1, inplace=True)
```

# 6. PCA 후 군집화 결과 확인

## 6.1. PCA 수행

- 누적 변동 비율을 확인하여 적절한 주성분 선정

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
ppc = pca.fit_transform(df_scale.drop(['BAS_ID'],axis=1))
df_pca = pd.DataFrame(ppc, columns=['PC1', 'PC2'])
df_pca
```

```python
sum(pca.explained_variance_ratio_)
```

```python
plt.title('주성분 관계 확인', size=20)
plt.xlabel('PC1', size=15)
plt.ylabel('PC2', size=15)
plt.scatter(df_pca['PC1'],df_pca['PC2'])
```

```python
df_pca = pd.concat([df_scale['BAS_ID'],df_pca],axis=1)
df_pca.head()
```

## 6.2. 군집화 수행

```python
affin = ['cosine','euclidean']
linkg = ['ward','complete','single','average']

df_pc5 = df_pscale[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=5, affinity=aff, linkage=lng).fit(df_pca.drop('BAS_ID',axis=1))
            col_name = '5pc_'+aff+'_'+lng
            df_pc5[col_name]=cluster.labels_
        except:
            continue
```

```python
km_cluster= KMeans(n_clusters=5).fit(df_pca.drop('BAS_ID',axis=1))
df_pc5['k_means_C5']=km_cluster.labels_

df_pc5.drop('BAS_AR', axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(2,4,sharex=True,sharey=True)
col_num = 1
for x in range(2):
    for y in range(4):
        try:
            ax[x][y].hist(df_pc5.iloc[:,col_num])
            ax[x][y].set_title(df_pc5.columns[col_num])
        except:
            continue
        col_num+=1
```

```python
df_pca['open_cos_avg'] = df_pc5['5pc_cosine_average']
df_pca['open_euc_ward'] = df_pc5['5pc_euclidean_ward']
```

## 6.3. 군집화 결과 시각화

```python
plt.title('영업기간 코사인유사도 주성분 군집화')
plt.scatter(df_pca.loc[df_pca['open_cos_avg']==0]['PC1'],
            df_pca.loc[df_pca['open_cos_avg']==0]['PC2'], color='red', label='0군집')
plt.scatter(df_pca.loc[df_pca['open_cos_avg']==1]['PC1'],
            df_pca.loc[df_pca['open_cos_avg']==1]['PC2'], color='green', label='1군집')
plt.scatter(df_pca.loc[df_pca['open_cos_avg']==2]['PC1'],
            df_pca.loc[df_pca['open_cos_avg']==2]['PC2'], color='blue', label='2군집')
plt.scatter(df_pca.loc[df_pca['open_cos_avg']==3]['PC1'],
            df_pca.loc[df_pca['open_cos_avg']==3]['PC2'], color='purple', label='3군집')
plt.scatter(df_pca.loc[df_pca['open_cos_avg']==4]['PC1'],
            df_pca.loc[df_pca['open_cos_avg']==4]['PC2'], color='grey', label='4군집')
plt.legend()
```

```python
plt.title('영업기간 유클리디안 주성분 군집화')
plt.scatter(df_pca.loc[df_pca['open_euc_ward']==0]['PC1'],
            df_pca.loc[df_pca['open_euc_ward']==0]['PC2'], color='red', label='0군집')
plt.scatter(df_pca.loc[df_pca['open_euc_ward']==1]['PC1'],
            df_pca.loc[df_pca['open_euc_ward']==1]['PC2'], color='green', label='1군집')
plt.scatter(df_pca.loc[df_pca['open_euc_ward']==2]['PC1'],
            df_pca.loc[df_pca['open_euc_ward']==2]['PC2'], color='blue', label='2군집')
plt.scatter(df_pca.loc[df_pca['open_euc_ward']==3]['PC1'],
            df_pca.loc[df_pca['open_euc_ward']==3]['PC2'], color='purple', label='3군집')
plt.scatter(df_pca.loc[df_pca['open_euc_ward']==4]['PC1'],
            df_pca.loc[df_pca['open_euc_ward']==4]['PC2'], color='grey', label='4군집')
plt.legend()
```
