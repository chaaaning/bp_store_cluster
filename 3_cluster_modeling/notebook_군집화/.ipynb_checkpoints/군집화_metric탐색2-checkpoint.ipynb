```python
import pandas as pd
import numpy as np
import os
import warnings

def row_num(x): return pd.set_option('display.max_rows',x)
def col_num(x): return pd.set_option('display.max_columns',x)

row_num(10)
col_num(10)

warnings.simplefilter('ignore')

os.chdir(r'C:\\Users\\user\\bupyeonggu\\store_cluster\\data_and_notebook\\정제데이터')
df = pd.read_csv('GWR 회귀계수.csv')
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering as agcluster
from sklearn.cluster import KMeans

%matplotlib inline
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(12,8)
plt.rcParams['figure.facecolor']='grey'
plt.rcParams['axes.grid']=True
```

```python
from sklearn.preprocessing import MinMaxScaler as MMS

tmp = pd.DataFrame(MMS().fit_transform(df.drop('BAS_ID',axis=1)), columns=df.drop('BAS_ID',axis=1).columns)
tmp2 = pd.concat([df['BAS_ID'],tmp],axis=1)
df = tmp2.copy()
df.drop(['x','y'],axis=1,inplace=True)
```

```python
affin = ['cosine','euclidean','manhattan']
linkg = ['ward','complete','single','average']

df_c5 = df[['BAS_ID','BAS_AR']]

for aff in affin:
    for lng in linkg:
        try:
            cluster = agcluster(n_clusters=5, affinity=aff, linkage=lng).fit(df)
            col_name = '5_'+aff+'_'+lng
            df_c5[col_name]=cluster.labels_
        except:
            continue

km_cluster = KMeans(n_clusters=5).fit(df)
df_c5['k_means_5']=km_cluster.labels_

df_c5.drop('BAS_AR', axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(3,4,sharex=True,sharey=True)
col_num = 1
for x in range(3):
    for y in range(4):
        try:
            ax[x][y].hist(df_c5.iloc[:,col_num])
            ax[x][y].set_title(df_c5.columns[col_num])
        except:
            continue
        col_num+=1
```
